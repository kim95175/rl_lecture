## DQN 설명
Q-Learning은 (State, Action) 쌍을 Table 형태로 나타내야 하기 때문에 State가 이산적이지 않거나, 혹은 State의 경우의 수가 너무 많은 경우 Q-Learning을 적용하기 어렵습니다.

DQN은 위 문제를 해결하기 위해, Q-value를 Q-table이 아닌 신경망을 통해 근사합니다.

Q-table 대신 신경망을 사용함으로써, State 형식이 보다 자유로워졌지만, 학습 과정에서 여러 문제가 생깁니다.

DQN 논문에서는 이를 Experience Replay, Target Network를 통해 해결합니다.


본 실습에서는,

1. Neural Net 구현 (필수)
1. Experience Replay 구현 (필수)
    - N-step Learning 구현 (선택) 
1. DQN Interaction, Training Loop 구현 (필수)
1. Double DQN 구현 (필수)

을 진행할 예정입니다.

## Replay Memory 구현
Numpy의 Array를 이용해 Circular Buffer로 구현하거나, Python에서 제공되는 Deque로 구현하시면 됩니다.

정답 코드에는 Circular Buffer로 구현되어 있습니다.

↓ 기본 정답 코드

```python
# "1"  with no N-step
class ReplayMemory:
    def __init__(
        self, 
        observation_shape: tuple = (),
        action_shape: tuple = (),
        buffer_size: int = 50000,
        num_steps: int = 1,
    ):
        self.observation_shape = observation_shape
        self.action_shape = action_shape
        self.buffer_size = buffer_size
        self.num_steps = num_steps
        
        # (s, a, r, s') + done 5개에 대해 각각 버퍼를 잡아줍니다.
        self.states = np.zeros(shape=(buffer_size, int(np.prod(observation_shape))), dtype=np.float32)
        self.actions = np.zeros(shape=(buffer_size, int(np.prod(action_shape))), dtype=np.int64)  # pytorch에서 indexing을 하려면 int64 자료형이 필요합니다.
        self.rewards = np.zeros(shape=(buffer_size, 1), dtype=np.float32)
        self.next_states = np.zeros(shape=(buffer_size, int(np.prod(observation_shape))), dtype=np.float32)
        self.dones = np.zeros(shape=(buffer_size, 1), dtype=np.bool)
        
        # index 변수로 Circular Buffer를 구현합니다.
        self.idx = 0
    
    
    def write(self, state, action, reward, next_state, done):
        self.states[self.idx % self.buffer_size] = state
        self.actions[self.idx % self.buffer_size] = action
        self.rewards[self.idx % self.buffer_size] = reward
        self.next_states[self.idx % self.buffer_size] = next_state
        self.dones[self.idx % self.buffer_size] = done
            
        self.idx += 1
    
    
    # 버퍼에서 Uniform Random으로 Transition들을 뽑습니다.
    def sample(self, num_samples: int = 1) -> Tuple[np.ndarray]:
        random_indices = np.random.choice(
            min(self.idx, self.buffer_size),
            size = min(num_samples, self.idx),
            replace = False
        )
        
        return (
            self.states[random_indices],
            self.actions[random_indices],
            self.rewards[random_indices],
            self.next_states[random_indices],
            self.dones[random_indices]
        )
```

↓ N-step 정답 코드

```python
# "1" with N-step
class ReplayMemory:
    def __init__(
        self, 
        observation_shape: tuple = (),
        action_shape: tuple = (),
        buffer_size: int = 50000,
        num_steps: int = 1,
    ):
        self.observation_shape = observation_shape
        self.action_shape = action_shape
        self.buffer_size = buffer_size
        self.num_steps = num_steps
        
        # (s, a, r, s') + done 5개에 대해 각각 버퍼를 잡아줍니다.
        self.states = np.zeros(shape=(buffer_size, int(np.prod(observation_shape))), dtype=np.float32)
        self.actions = np.zeros(shape=(buffer_size, int(np.prod(action_shape))), dtype=np.int64)  # pytorch에서 indexing을 하려면 int64 자료형이 필요합니다.
        self.rewards = np.zeros(shape=(buffer_size, 1), dtype=np.float32)
        self.next_states = np.zeros(shape=(buffer_size, int(np.prod(observation_shape))), dtype=np.float32)
        self.dones = np.zeros(shape=(buffer_size, 1), dtype=np.bool_)
        
        # index 변수로 Circular Buffer를 구현합니다.
        self.idx = 0
        
        # N-step 구현을 위한 버퍼들을 준비해줍니다.
        self.n_states = deque(maxlen=self.num_steps + 1)
        self.n_actions = deque(maxlen=self.num_steps + 1)
        self.n_rewards = deque(maxlen=self.num_steps + 1)
    
    
    # @property를 붙이면 호출 시 ()가 없어도 됩니다. 자주 참조하는 특성에 사용하면
    # num_steps와 같이 민감한 변수를 직접 호출하지 않을 수 있습니다. (실수로 인한 대입을 막습니다)
    @property   
    def is_multistep(self):
        return self.num_steps > 1
    
    
    # 내장함수를 구현해, len() 함수를 통해 버퍼의 크기를 알 수 있습니다.
    def __len__(self):
        return min(self.idx, self.buffer_size)
    
    
    def __write(self, state, action, reward, next_state, done):
        self.states[self.idx % self.buffer_size] = state
        self.actions[self.idx % self.buffer_size] = action
        self.rewards[self.idx % self.buffer_size] = reward
        self.next_states[self.idx % self.buffer_size] = next_state
        self.dones[self.idx % self.buffer_size] = done
            
        self.idx += 1
        
    
    
    def write(self, state, action, reward, next_state, done):
        if not self.is_multistep:
            self.__write(state, action, reward, next_state, done)
        else:
            self.n_states.append(state)
            self.n_actions.append(action)
            self.n_rewards.append(reward)
            
            if done:
                self.multistep_finish()
            elif len(self.n_states) == self.num_steps + 1:  # N + 1개의 State는 N개의 transitions를 (steps) 의미합니다.
                self.__write(self.n_states[0], self.n_actions[0], sum(self.n_rewards), self.n_states[-1], done)
    
    
    # 에피소드가 끝났을 때, (done이 True일 때) Multistep 버퍼를 비우며 모든 Step이 Replay에 기록될 수 있도록 합니다.
    def multistep_finish(self):
        while len(self.n_states) > 1:
            self.__write(self.n_states.popleft(), self.n_actions.popleft(), sum(self.n_rewards), self.n_states[-1], True)
            self.n_rewards.popleft()
    
    
    # 버퍼에서 Uniform Random으로 Transition들을 뽑습니다.
    def sample(self, num_samples: int = 1) -> Tuple[np.ndarray]:
        random_indices = np.random.choice(
            min(self.idx, self.buffer_size),
            size = min(num_samples, self.idx),
            replace = False
        )
        
        return (
            self.states[random_indices],
            self.actions[random_indices],
            self.rewards[random_indices],
            self.next_states[random_indices],
            self.dones[random_indices]
        )
```

## NN 구현
단순한 MLP를 이용합니다. DQN에서는 Observation을 입력받아, Q-value를 출력합니다.

↓ 정답 코드 (예제)

```python
# "2"  "3"
class SimpleMLP(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.linear1 = nn.Linear(input_size, 64)
        self.linear2 = nn.Linear(64, 128)
        self.linear3 = nn.Linear(128, 128)
        self.linear4 = nn.Linear(128, 64)
        self.linear5 = nn.Linear(64, output_size)
    
    
    def forward(self, x):
        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        x = F.relu(self.linear3(x))
        x = F.relu(self.linear4(x))
        
        return self.linear5(x)
```

## DQN 학습 코드
Target Network와 Epsilon-Greedy를 통해 학습하는 DQN의 메인 로직 구현입니다.

↓ 정답 코드 (예제)

```python
class DQNTrainer:
    def __init__(
        self,
        env: gym.Env,
        use_target: bool = True,
        optim_cls: type = torch.optim.Adam,
        optim_kwargs: dict = {'lr': 0.0005},
        loss_fn: nn.modules.loss = torch.nn.SmoothL1Loss(),
        eps_start: float = 0.3,
        eps_end: float = 0.01,
        eps_decay: float = 0.9995,  
        discount_rate: float = 0.98,
        n_steps: int = 1,
        double: bool = False,
        buffer_size: int = 50000,
        batch_size: int = 128,
        learning_starts: int = 512,
        train_freq: int = 1,
        target_update_freq: int = 2048,
        device="cuda"
    ):
        self.env = env
        self.use_target = use_target
        self.loss_fn = loss_fn
        self.epsilon = eps_start
        self.eps_end = eps_end
        self.eps_decay = eps_decay
        self.discount_rate = discount_rate
        self.n_steps = n_steps
        self.double = double
        self.buffer_size = buffer_size
        self.batch_size = batch_size
        self.learning_starts = learning_starts
        self.train_freq = train_freq
        self.target_update_freq = target_update_freq
        self.device = device
        
        # 학습 과정 기록, 출력을 위한 변수들입니다.
        self.steps = 0
        self.episodes = 0
        
        self.alpha = 0.97
        self.reward_ema = 0.0
        self.episode_reward_ema = 0.0
        self.episode_length_ema = 0.0
        
        
        # "1". Initialize replay memory D to capacity N
        self.replay_memory = ReplayMemory(
            self.env.observation_space.shape,
            self.env.action_space.shape,
            self.buffer_size,
            self.n_steps
        )
        
        # "2". Initialize action-value function Q with random weights θ
        self.online_net = SimpleMLP(
            int(np.prod(self.env.observation_space.shape)),
            self.env.action_space.n
        ).to(self.device)
        
        # "3". Initialize target action-value function Q_hat with weights θ^- = θ
        self.target_net = SimpleMLP(
            int(np.prod(self.env.observation_space.shape)),
            self.env.action_space.n
        ).to(self.device)
        self.target_net.load_state_dict(self.online_net.state_dict())
        
        self.optimizer = optim_cls(self.online_net.parameters(), **optim_kwargs)
        
    
    def train(self, num_train_steps: int):
        begin_steps = self.steps
        
        episode_rewards = []
        episode_lengths = []
        episode_steps = []
        
        # 원래 pseudocode에서는 episode 횟수로 루프하지만, 여기서는 실질적인 환경과의 Interaction 횟수인 Step 수를 기준으로 잡았습니다.
        # "4". For episode = 1, M do
        while self.steps - begin_steps < num_train_steps:
            # "5". Initialize sequence s_1 = {x_1} and preprocessed sequence φ_1 = φ(s_1)
            done, ob = False, env.reset()
            episode_reward = 0.0
            episode_length = 0
            
            # 에피소드가 끝날 때 까지 루프
            # " 5". For t = 1, T do
            while not done:
                
                # Epsilon-Greedy로 Action 추출
                epsilon = self.epsilon
                self.update_epsilon()
                
                # "6". With probability ε select a random action a_t
                if np.random.random() < epsilon:
                    action = self.env.action_space.sample()
                # "7". otherwise select a_t = argmax_a Q(φ(s_t), a; θ)
                else:
                    action = torch.argmax(self.online_net.forward(torch.Tensor(ob).to(self.device))).item()
                
                # "8". Execute action a_t in emulator and observe reward r_t and image x_(t+1)
                next_ob, reward, done, info = env.step(action)
                
                # "9". Store transition (φ_t, a_t, r_t, φ_(t + 1)) in D.  지금은 φ_t와 x_t는 observation과 같습니다. Done도 함께 저장해야 합니다.
                self.replay_memory.write(ob, action, reward, next_ob, done)
                ob = next_ob  # 직전 State를 다시 저장해줍니다.
                
                # 정해진 step 수 마다 Q-Network를 학습시킵니다.
                if self.steps > self.learning_starts and self.steps % self.train_freq == 0:
                    self.update_network()
                
                # 정해진 step 수 마다 Target Network를 업데이트합니다.
                # "13". Every C steps reset Q_hat = Q
                if self.steps % self.target_update_freq == 0:
                    self.update_target()
            
                
                self.steps += 1
                episode_reward += reward
                episode_length += 1
                self.reward_ema = self.alpha * self.reward_ema + (1.0 - self.alpha) * reward
                
            print(f"\r Steps: {self.steps - begin_steps}/{num_train_steps}, Epsilon: {epsilon:.4f},"\
                    + f" Avg. Reward: {self.reward_ema:.2f}, Avg. Epi. Reward: {self.episode_reward_ema:.2f},"\
                    + f" Last Epi. Reward: {episode_reward}     ", end="", flush=True)
            
            self.episode_reward_ema = self.alpha * self.episode_reward_ema + (1.0 - self.alpha) * episode_reward
            self.episode_length_ema = self.alpha * self.episode_length_ema + (1.0 - self.alpha) * episode_length
            episode_rewards.append(episode_reward)
            episode_lengths.append(episode_length)
            episode_steps.append(self.steps)
            
        return episode_rewards, episode_lengths, episode_steps
    
    
    def update_network(self):
        self.optimizer.zero_grad()
        
        # Experience Replay에서 랜덤하게 Experiences를 뽑아옵니다.
        # "10". Sample random minibatch of transitions (φ_j, a_j, r_j, φ_(j + 1)) from D
        state, action, reward, next_state, done = self.replay_memory.sample(self.batch_size)
        
        state = torch.FloatTensor(state).to(self.device)
        action = torch.LongTensor(action).to(self.device)
        reward = torch.FloatTensor(reward).to(self.device)
        next_state = torch.FloatTensor(next_state).to(self.device)
        done = torch.FloatTensor(done).to(self.device)
        
        # "11". set y_j = {...}
        with torch.no_grad():
            if not self.double:  # "11"-1. y_j = r_j + γ * max_a' Q_hat(φ_j, a'; θ^-)
                if self.use_target:
                    target = self.discount_rate * torch.max(self.target_net(next_state), dim=-1, keepdim=True)[0] * (1 - done) + reward
                else:
                    target = self.discount_rate * torch.max(self.online_net(next_state), dim=-1, keepdim=True)[0] * (1 - done) + reward

            else:  # " Double". y_j = r_j + γ Q_hat(φ_j, argmax_a Q(φ_j, a; θ); θ^-) 
                next_action = torch.argmax(self.online_net(next_state), dim=-1, keepdim=True)
                target = self.discount_rate * self.target_net(next_state).gather(1, next_action) * (1 - done) + reward
        
        # "12". Perform a gradient step on (y_j - Q(φ_j, a_j; θ))^2 with respect to the network parmeters θ.
        pred = self.online_net(state).gather(1, action)
        loss = self.loss_fn(pred, target)
        loss.backward()
        
        for param in self.online_net.parameters():
            param.grad.data.clamp_(-1, 1)
        
        self.optimizer.step()
    
    
    def predict(self, ob):
        return torch.argmax(self.online_net.forward(torch.Tensor(ob).to(self.device))).item()
        
        
    def update_epsilon(self):
        self.epsilon = max(self.epsilon * self.eps_decay, self.eps_end)
    
    
    def update_target(self):
        self.target_net.load_state_dict(self.online_net.state_dict())
```

### DQN 학습 진행

```python
%%time
env = gym.make("CartPole-v1")
plain_trainer = DQNTrainer(env, device="cuda", double=False, n_steps=1)
plain_result = plain_trainer.train(50000)
```

```python
%%time
env = gym.make("CartPole-v1")
double_trainer = DQNTrainer(env, device="cuda", double=True, n_steps=1)
double_result = double_trainer.train(50000)
```

```python
%%time
env = gym.make("CartPole-v1")
dbl_multi_trainer = DQNTrainer(env, device="cuda", double=True, n_steps=8)
dbl_multi_result = dbl_multi_trainer.train(50000)
```

```python
%%time
env = gym.make("CartPole-v1")
no_replay_trainer = DQNTrainer(env, device="cuda", batch_size=128, buffer_size=128, double=False, n_steps=1)
no_replay_result = no_replay_trainer.train(50000)
```

```python
%%time
env = gym.make("CartPole-v1")
no_target_trainer = DQNTrainer(env, device="cuda", use_target=False, double=False, n_steps=1)
no_target_result = no_target_trainer.train(50000)
```

# 학습 결과 시각화
학습 과정에 대한 그래프와, 학습이 완료된 에이전트의 능력을 동영상으로 확인해봅니다.

```python
def plot_with_exponential_averaging(x, y, label, alpha):
    y_ema = [y[0]] 
    for y_i in y[1:]:
        y_ema.append(y_ema[-1] * alpha + y_i * (1 - alpha))
    
    p = plt.plot(x, y_ema, label=label)
    
    plt.plot(x, y, color=p[0].get_color(), alpha=0.2)
    

def plot_dqn_train_result(result, label="", alpha=0.95):
    plot_with_exponential_averaging(result[2], result[0], label, alpha)
```

### 학습 과정

```python
plt.cla()
plot_dqn_train_result(plain_result, label="Vanilla DQN", alpha=0.9)
plot_dqn_train_result(double_result, label="Double DQN", alpha=0.9)
plot_dqn_train_result(dbl_multi_result, label="Double + 8-Steps", alpha=0.9)
plot_dqn_train_result(no_replay_result, label="No Replay DQN", alpha=0.9)
plot_dqn_train_result(no_target_result, label="No Target DQN", alpha=0.9)

plt.axhline(y=500, color='grey', linestyle='-')
plt.xlabel("steps")
plt.ylabel("Episode Reward")

plt.legend()
plt.title("Training Comparison between Techniques for DQN")
plt.show()
```
